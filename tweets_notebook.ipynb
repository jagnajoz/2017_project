```python
#import modules
import re
import matplotlib.pyplot as plt
import emoji_list
import nltk
#For the first run download nltk corpora
#nltk.download
from nltk.corpus import twitter_samples
from nltk.tag import pos_tag_sents
from wordcloud import WordCloud, STOPWORDS
```

```python
#save and tokenize
ntweets = nltk.corpus.twitter_samples.strings('negative_tweets.json')
ntweets_tok = nltk.corpus.twitter_samples.tokenized('negative_tweets.json')
ptweets = nltk.corpus.twitter_samples.strings('positive_tweets.json')
ptweets_tok = nltk.corpus.twitter_samples.tokenized('positive_tweets.json')

#tag with parts of speech
ntweets_tagged = pos_tag_sents(ntweets_tok)
ptweets_tagged = pos_tag_sents(ptweets_tok)
```

```python
#define POS and emojis
emolist = emoji_list.all_emoji + [ ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3', ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';(']
```

```python
#list profanities
#fword = r'fuc+k+'
#sword = r'shi+e*t+'
#dword = r'dam+n+'

nprofan = []
pprofan = []
for tweet in ntweets_tok:
    for word in tweet:
        if re.search('fuc+k+', word) or re.search('shi+e*t+', word) or re.search('dam+n+', word):
            nprofan.append(word)
for tweet in ptweets_tok:
    for word in tweet:
        if re.search('fuc+k+', word) or re.search('shi+e*t+', word) or re.search('dam+n+', word):
            pprofan.append(word)
```

```python
#plot profanities
pprofanstring = ''
for i in pprofan:
    pprofanstring += i + ' '

    
nprofanstring = ''
for i in nprofan:
    nprofanstring += i + ' '
print('Swearwords in POSITIVE TWEETS')
wordcloud = WordCloud(font_path='/Library/Fonts/Verdana.ttf',
                      stopwords = {' '} # set or space-separated string
                      ).generate(pprofanstring)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
print('Swearwords in NEGATIVE TWEETS')
wordcloud = WordCloud(font_path='/Library/Fonts/Verdana.ttf',
                      stopwords = {' '} # set or space-separated string
                      ).generate(nprofanstring)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
#count POS

nemo_count = 0
nadj_count = 0
nnoun_count = 0
nadv_count = 0
nverb_count = 0
nnounlist = []
nverblist = []
    
for tweet in ntweets_tagged:
    for t in tweet:
        content = t[0]
        tag = t[1]          
        if content in emolist:
            nemo_count += 1
        elif tag == 'NN' or tag == 'NNS' or tag == 'NNP':
            nnoun_count += 1
            nnounlist.append(content)
        elif tag == 'RB':
            nadv_count += 1
        elif tag == 'JJ':
            nadj_count += 1
        elif tag == 'VB' or tag == 'VBP':
            nverb_count += 1
            nverblist.append(content)
        

pemo_count = 0
padj_count = 0
pnoun_count = 0
padv_count = 0
pverb_count = 0
pnounlist = []
pverblist = []

for tweet in ptweets_tagged:
    for t in tweet:
        content = t[0]
        tag = t[1]          
        if content in emolist:
            pemo_count += 1
        elif tag == 'NN' or tag == 'NNS' or tag == 'NNP':
            pnoun_count += 1
            pnounlist.append(content)
        elif tag == 'RB':
            padv_count += 1
        elif tag == 'JJ':
            padj_count += 1
        elif tag == 'VB' or tag == 'VBP':
            pverblist.append(content)
            pverb_count += 1
pverblist
```

```python
#plot the POS pie chart
#positive
print("Parts of speech in POSITIVE TWEETS")
labels = 'Nouns', 'Verbs', 'Adjectives', 'Adverbs', 'Emoji'
sizes = [pnoun_count, pverb_count, padj_count, padv_count, pemo_count]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

```

```python
#negative
print("Parts of speech in NEGATIVE TWEETS")
labels = 'Nouns', 'Verbs', 'Adjectives', 'Adverbs', 'Emoji'
sizes = [nnoun_count, nverb_count, nadj_count, nadv_count, nemo_count]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
